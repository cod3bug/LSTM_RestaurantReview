## DATA PREPROCESSING

### Importing dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('Restaurant_Reviews.tsv', delimiter = '\t', quoting = 3)
```

```python
X = df.drop("Liked",axis=1)
```

```python
y = df['Liked']
```

```python
X.shape
```

```python
y.shape
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

### Initializing dictionary size

```python
voc_size = 10000
```

```python
reviews = X.copy()
```

```python
reviews['Review'][1]
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(reviews)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', reviews['Review'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

### Generating the architecture of LSTM 

```python
from tensorflow.keras.layers import Dropout
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

### Training the model

```python
model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred=model.predict_classes(X_test)
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### Predicting result on normal text input 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features = 1000)
cv.fit(corpus)
```

```python
pred = ["The food was bad"]
vect = cv.transform(pred).toarray()
```

```python
if model.predict(vect)[0] >= 0.5:
    print("Positive")
else:
    print("Negative")
```
